```python
from datetime import datetime
from google.cloud import bigquery
from google.oauth2 import service_account
from google.cloud import bigquery
from google.cloud.exceptions import NotFound
import pandas as pd
import yahoo_fin.stock_info as si
from utils.utils_bigquery import *
from datetime import *
import numpy as np
```

```python
key_path = key_path
project = project_id

# Sources
table = 'silver_indicators'
dataset = 'silver'
table_conca = f'{project}.{dataset}.{table}'

# Loading tables
table_to_save = 'gold_dim_candles'
dataset_to_save = 'gold'
table_conca_to_save = f'{project}.{dataset_to_save}.{table_to_save}'
```

```python
# Conectamos con Bigquery
bigquery = BigQueryUtils(key_path)
```

```python
df = bigquery.run_query(
    f"""
    SELECT
        CDL_DOJI,
        CDL_HAMMER,
        CDL_MORNING_STAR,
        CDL_SHOOTING_STAR,
        CDL_ENGULFING,
        CDL_LONGLINE
    FROM {project}.{dataset}.{table} AS can
    LIMIT 10
    """
)
df
```

```python
# Transform the DataFrame to long format (unpivot)
df_unpivot = df.melt(var_name='candles', value_name='results')

# Drop the 'results' column
df_unpivot = df_unpivot.drop(columns='results')

# Get unique values from the 'technical_indicators' column and convert to DataFrame
unique_candels = pd.DataFrame(df_unpivot['candles'].unique(), columns=['candles'])

# Display the result
unique_candels
```

```python
bigquery.save_dataframe(unique_candels, project, dataset_to_save, table_to_save, if_exists='replace', schema=None)
```
