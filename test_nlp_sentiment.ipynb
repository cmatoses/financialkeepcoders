```python
from stocksent import Sentiment
stocks = Sentiment(['AAPL','TSLA','AMZN'])
sentiment_score = stocks.get_dataframe(days=1825) # Get the headlines for the past X days.

sentiment_score.head()  # Returns a DataFrame with headlines, source and sentiment scores.
```

```python
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline

# Inicializar el tokenizer y el modelo FinBERT
tokenizer = BertTokenizer.from_pretrained("yiyanghkust/finbert-tone")
model = BertForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone")
nlp = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# Suponiendo que 'sentiment_headline' es un DataFrame con columnas 'date', 'ticker', 'headline'
# Aplicar el análisis de sentimiento a cada headline y añadir los resultados en una nueva columna

nlp()
# Mostrar el DataFrame con la nueva columna
sentiment_score.head(10)
```

```python
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline

# Inicializar el tokenizer y el modelo FinBERT
tokenizer = BertTokenizer.from_pretrained("yiyanghkust/finbert-tone")
model = BertForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone")
nlp = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# Suponiendo que 'sentiment_headline' es un DataFrame con columnas 'date', 'ticker', 'headline'
# Aplicar el análisis de sentimiento a cada headline y añadir los resultados en una nueva columna
sentiment_score['sentiment'] = sentiment_score['headline'].apply(lambda x: nlp(x)[0]['label'])

# Mostrar el DataFrame con la nueva columna
sentiment_score.head(10)
```
