```python
historical_datas = []

# Recorrer los smbolos y obtener datos
for symbol in symbols:
    try:
        # Obtener datos histricos
        data = si.get_data(symbol)
        # Agregar el smbolo como una columna en los datos histricos
        data['Symbol'] = symbol
        # Aadir los datos al diccionario
        historical_datas.append(data)
        print(f"Datos obtenidos para {symbol}.")
    except Exception as e:
        print(f"Error al obtener datos para {symbol}: {e}")

# Mostrar las primeras filas del DataFrame resultante
print(data.head())
```
