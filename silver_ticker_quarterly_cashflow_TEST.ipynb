```python
# Function to calculate the percentage of empty cells for each column
def percentage_empty_cells_per_column(df):
    
    # Calculate the percentage of empty cells for each column
    empty_percentage = df.isna().mean() * 100
    return empty_percentage

# Function to remove columns where the percentage of empty cells is less than a given threshold
def remove_columns_with_low_empty_percentage(df, threshold):
    # Get the percentage of empty cells for each column using the first function
    empty_percentage = percentage_empty_cells_per_column(df)
    
    # Select columns with empty percentage greater than or equal to the threshold
    columns_to_keep = empty_percentage[empty_percentage >= threshold].index
    
    # Return DataFrame with the selected columns
    filtered_df = df[columns_to_keep]
    return filtered_df

threshold = 50  # Threshold percentage

df_vacias = percentage_empty_cells_per_column(df)
df_vacias
```

```python
def remove_columns_with_low_empty_percentage(df, threshold):
    # Calculate the percentage of empty cells for each column
    empty_percentage = df.isna().mean() * 100
    
    # Select columns with empty percentage greater than or equal to the threshold
    columns_to_keep = empty_percentage[empty_percentage <= threshold].index
    
    # Return DataFrame with the selected columns
    return df[columns_to_keep]

threshold = 30  # Threshold percentage

filtered_df = remove_columns_with_low_empty_percentage(df, threshold)
filtered_df.info()
```

```python
try:
    # Filtramos solamente los nuevos registros
    df_incremental = bigquery.select_for_incremental(id='id', table=table_conca_to_save, new_df=df)

    if not df_incremental.empty:
        # Guardamos los datos en Bigquery
        bigquery.save_dataframe(df_incremental, project, dataset_to_save, table_to_save, if_exists='append', schema=None)
        print(f'New records loaded')
    else:
        print('No new records to load.')

# En el caso de no tener datos en Bigquery, guardamos todo el df
except Exception as e:
    bigquery.save_dataframe(df_incremental, project, dataset_to_save, table_to_save, if_exists='replace', schema=None)
    print('New data persisted')
    print(f'Exception encountered: {e}')
```
