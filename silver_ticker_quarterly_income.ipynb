```python
from datetime import datetime
from google.cloud import bigquery
from google.oauth2 import service_account
from google.cloud import bigquery
from google.cloud.exceptions import NotFound
import pandas as pd
import numpy as np
import yahoo_fin.stock_info as si
from utils.utils_bigquery import *
from datetime import *
```

```python
key_path = key_path
project = project_id

# Sources
table = 'bronze_ticker_quarterly_income'
dataset = 'bronze'
table_conca = f'{project}.{dataset}.{table}'

# Loading tables
table_to_save = 'silver_ticker_quarterly_income'
dataset_to_save = 'silver'
table_conca_to_save = f'{project}.{dataset_to_save}.{table_to_save}'
```

```python
# Conectamos con Bigquery
bigquery = BigQueryUtils(key_path)
```

```python
df = bigquery.run_query(
    f"""
    SELECT
        *
    FROM {project}.{dataset}.{table}
    """
)
df
```

```python
df = df.rename(columns={"index": "assets"})
df
```

```python
# Transformar el DataFrame a formato largo (long format)
df_unpivot = df.melt(id_vars=['ticker', 'assets'], var_name='ReportDate', value_name='Results')

# Convertir las fechas a datetime
df_unpivot['ReportDate'] = pd.to_datetime(df_unpivot['ReportDate'], format='%Y-%m-%d')

df_unpivot
```

```python
df_unpivot = df_unpivot.dropna()
df_unpivot
```

```python
# Definir los trimestres con meses de presentación alternativos
def get_quarter(date):
    if date.month in [3, 4, 5]:
        return (date.year, 1)
    elif date.month in [6, 7, 8]:
        return (date.year, 2)
    elif date.month in [9, 10, 11]:
        return (date.year, 3)
    elif date.month in [12, 1, 2]:
        return (date.year if date.month == 12 else date.year - 1, 4)

df_unpivot['Quarter'] = df_unpivot['ReportDate'].apply(get_quarter)
df_unpivot
```

```python
# Crear un DataFrame vacío para los datos diarios
daily_data = []

# Rellenar los datos diarios con los resultados trimestrales
for _, row in df_unpivot.iterrows():
    year, quarter = row['Quarter']
    if quarter == 1:
        start_date = f"{year}-01-01"
        end_date = f"{year}-03-31"
    elif quarter == 2:
        start_date = f"{year}-04-01"
        end_date = f"{year}-06-30"
    elif quarter == 3:
        start_date = f"{year}-07-01"
        end_date = f"{year}-09-30"
    elif quarter == 4:
        start_date = f"{year-1}-10-01"
        end_date = f"{year-1}-12-31"

    date_range = pd.date_range(start=start_date, end=end_date, freq='D')
    for date in date_range:
        daily_data.append({'ticker': row['ticker'], 'assets': row['assets'], 'date': date, 'Results': row['Results']})

# Crear un DataFrame con los datos diarios
daily_df = pd.DataFrame(daily_data)

daily_df
```

```python
# Pivotar el DataFrame
pivoted_df = daily_df.pivot_table(index=['ticker', 'date'], columns='assets', values='Results')
pivoted_df = pivoted_df.reset_index()
pivoted_df.columns.name = None  # Esto elimina el nombre de la columna del indice

pivoted_df
```

```python
pivoted_df.info()
```

```python
# Fields that make up the ID
id_fields = ['ticker', 'date']

# Apply the function to the DataFrame to create the 'id' column
pivoted_df['id'] = pivoted_df.apply(generate_id, axis=1, fields=id_fields)

pivoted_df
```

```python
try:
    # Filtramos solamente los nuevos registros
    df_incremental = bigquery.select_for_incremental(id='id', table=table_conca_to_save, new_df=pivoted_df)

    if not df_incremental.empty:
        # Guardamos los datos en Bigquery
        bigquery.save_dataframe(df_incremental, project, dataset_to_save, table_to_save, if_exists='append', schema=None)
        print(f'New records loaded')
    else:
        print('No new records to load.')

# En el caso de no tener datos en Bigquery, guardamos todo el df
except Exception as e:
    bigquery.save_dataframe(pivoted_df, project, dataset_to_save, table_to_save, if_exists='replace', schema=None)
    print('New data persisted')
    print(f'Exception encountered: {e}')
```
